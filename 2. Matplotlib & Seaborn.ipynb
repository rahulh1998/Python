# Simple Plot

```python
x_1 = np.linspace(0,10,100)
y_1 = x_1**2
plt.plot(x_1,y_1)
plt.title("Normal line")
plt.xlabel('Deciles')
plt.ylabel('Percentage')
plt.show()
```

# Subplots

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
plt.show()
```

# Figure

```python
# figsize in inches
fig1 = plt.figure(figsize=(4,4),dpi = 100)
ax_1 = fig1.add_axes([0.1,0.1,0.9,0.9])#lbwh
ax_1.set_xlabel('Days')
ax_1.set_ylabel('Days Squared')
ax_1.plot(x_1,y_1,label = 'X/X^2')
ax_1.plot(y_1,x_1,label = 'X^2/X')
ax_1.legend(loc = 0)#0 = best location

ax2 = fig1.add_axes([0.45,0.45,0.3,0.3])
ax2.plot(x_1,y_1,'r',label = 'X/X^2')
ax2.plot(y_1,x_1,'g',label = 'X^2/X')
#ax2.legend(loc = 0)
#ax2.set_xlim([0,200])
#ax2.set_axis_off()
#ax_1.set_axis_off()

ax_1.text(40,100,'Message')

plt.show()
```

# Subplots

```python
fig2 , axes_2 = plt.subplots(figsize = (10,5),ncols=3,nrows = 1)
plt.tight_layout()
axes_2[1].set_xlabel('X')
axes_2[1].set_ylabel('Y')
axes_2[1].plot(y_1,x_1,'g')

axes_2[0].set_xlabel('X')
axes_2[0].set_ylabel('Y')
axes_2[0].plot(y_1,x_1,'r')

axes_2[2].set_xlabel('X')
axes_2[2].set_ylabel('Y')
axes_2[2].plot(y_1,x_1,'b')
plt.show()
```

```python
fig3 = plt.figure(figsize=(4,4),dpi=100)
axes1 = fig3.add_axes([0,0,1,1])
axes1.set_xlabel('Days')
axes1.set_ylabel('Days Squared')
axes1.plot(x_1,y_1,label = 'X/X^2',color = 'navy',lw = 2,ls ='--',alpha = 0.5)
axes1.plot(y_1,x_1,label = 'X^2/X',marker = 'o',markerfacecolor= 'y',markeredgecolor = 'y',markersize = 2,markeredgewidth =4)
axes1.set_title('Customized Figure')
axes1.legend(loc = 0)
axes1.set_facecolor('gray')
axes1.grid()
plt.show()
```

# Pandas Visualization

```python
df = pd.read_csv('datasets/bank.csv',sep=';')
df.head()
```

```python
day = np.array(df['day'].sort_values())
bal = np.array(df['balance'])

fig4 = plt.figure(figsize=(5,5))
ax4 = fig4.add_axes([0,0,1,1])

ax4.set_title('Balance per Day')
ax4.set_xlabel("Days")
ax4.set_ylabel('Balance')
ax4.bar(day,bal)
ax4.set_xlim([0,32])
ax4.set_ylim([0,75000])
ax4.annotate('Salary Credit', xy=(22,72000))
plt.show()
```

```python
pt = df.pivot_table(values='balance',index='day',aggfunc=[sum,np.mean,np.min,np.max])
pt.columns = ['Sum_Bal','Mean_Bal','Min_Bal','Max_Bal']
pt.head()
```

```python
fig5 = plt.figure(figsize=(10,4),dpi = 100)
ax5 = fig5.add_axes([0,0,1,1])
ax5.set_title('Balance distribution over Days')
ax5.set_xlabel('Days')
ax5.plot(pt,label=pt.columns)
ax5.legend()
plt.show()
```

```python
df2 = pd.read_csv('datasets/bank.csv',sep = ';')
df2.head()
```

# Bar Plots

```python
sns.barplot(x = 'job',y='balance',data = df2,hue = 'default',estimator=np.mean)
plt.xticks(rotation = 90)
plt.legend(loc=4)
plt.show()
```

```python
sns.histplot(x = 'job',data = df2,hue = 'default',kde = True)
plt.xticks(rotation = 90)
plt.show()
```

# Countplot

```python
sns.countplot(df2['education'],hue = df2['default'])
```

# Box plots

```python
plt.figure(figsize= (10,8))
sns.boxplot(df2['education'],df2['balance'],hue = df2['default'])
plt.ylim([-2000,10000])
plt.show()
```

# Violin Plot

```python
plt.figure(figsize= (10,8))
sns.violinplot(df2['education'],df2['balance'],hue = df2['default'])
plt.ylim([-2000,10000])
plt.show()
```

```python
plt.figure(figsize= (10,8))
sns.violinplot(df2['education'],df2['balance'],hue = df2['default'], split = True)
plt.ylim([-2000,10000])
plt.show()
```

# Stripplot

```python
plt.figure(figsize= (10,8))
sns.stripplot(df2['education'],df2['balance'],hue = df2['default'],jitter = True, dodge=True)
plt.ylim([-2000,10000])
plt.show()
```
