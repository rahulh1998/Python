```python
# place the list of categories Series inside stacked bar chart
# for legends either pass list of labels inside legends() or inside plot func labels = labels_list

l = []
legend = []
for i in df['Genre'].unique():
    l.append(df[df['Genre']==i].Budget)
    legend.append(i)

plt.figure(figsize=(12,8))
plt.hist(l,stacked = True,bins=30,label=legend)
plt.legend()
plt.show()
```

```python
# place the list of categories Series inside stacked bar chart
# for legends either pass list of labels inside legends() or inside plot func labels = labels_list

l = []
legend = []
for i in df['Genre'].unique():
    l.append(df[df['Genre']==i].Budget)
    legend.append(i)

plt.figure(figsize=(12,8))
plt.hist(l,stacked = False,bins=10,label=legend)
plt.legend()
plt.show()
```

 <h1 style = 'color:Gray;font-size:45px;text-align:center'> Subplots</h1>

```python
plt.figure(figsize=(20,10))
for i,j in enumerate(df['Genre'].unique()):
    plt.subplot(3,3,i+1,xticks = [])
    plt.hist(df[df['Genre'] == j]['Budget'],color= 'gray',histtype='barstacked')
    plt.title(j)
plt.show()
```
